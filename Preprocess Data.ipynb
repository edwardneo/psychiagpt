```python
import numpy as np
import pandas as pd
import re
import nltk
import spacy
import string
import os
from nltk import sent_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
def separate_into_sentences(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        
    sentences = sent_tokenize(text)
    return sentences
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
```

```python
path = "./anxiety.txt"
sentences = separate_into_sentences(path)
a_df = pd.DataFrame({'Sentences': sentences})
a_df["text_lower"] = a_df["Sentences"].str.lower()
PUNCT_TO_REMOVE = string.punctuation
a_df["text_wo_punct"] = a_df["text_lower"].apply(lambda text: remove_punctuation(text))
STOPWORDS = set(stopwords.words('english'))
a_df["text_wo_stop"] = a_df["text_wo_punct"].apply(lambda text: remove_stopwords(text))
a_df.head()
```

```python
path = "./depression.txt"
sentences = separate_into_sentences(path)
d_df = pd.DataFrame({'Sentences': sentences})
d_df["text_lower"] = d_df["Sentences"].str.lower()
PUNCT_TO_REMOVE = string.punctuation
d_df["text_wo_punct"] = d_df["text_lower"].apply(lambda text: remove_punctuation(text))
STOPWORDS = set(stopwords.words('english'))
d_df["text_wo_stop"] = d_df["text_wo_punct"].apply(lambda text: remove_stopwords(text))
d_df.head()
```

```python
# Create a new DataFrame with a label column
new_df = pd.DataFrame({
    'data_column': a_df["text_wo_stop"],
    'label': 1
})
print(len(new_df))

# Print the new DataFrame
additional_rows = pd.DataFrame({
    'data_column': d_df["text_wo_stop"],
    'label': 0
})
print(len(additional_rows))


# Concatenate the original 'new_df' and the additional_rows
new_df = pd.concat([new_df, additional_rows], ignore_index=True)
print(new_df)
```

```python
csv_file_path = './DSM.csv'

# Save the DataFrame to a CSV file
new_df.to_csv(csv_file_path, index=False)

print(f'DataFrame saved to {csv_file_path}')
```
