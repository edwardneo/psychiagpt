```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_curves(runs, loss_type, filename, names=None, title=None, xticks=[0, 2, 4, 6, 8, 10], steps_per_ticks=50):
  if names == None:
    names = runs
  if title == None:
    title = loss_type

  losses = [np.load(run + "/" + loss_type + ".npy") for run in runs]

  for loss, name in zip(losses, names):
    plt.plot(loss, label=name)

  plt.title(title)
  plt.xlabel("Steps")
  plt.ylabel("Cross Entropy Loss")
  plt.xticks(xticks, np.array(xticks) * steps_per_ticks)
  plt.legend()
  plt.savefig(filename)
  plt.show()
```

```python
plot_curves(
    runs=["lora8", "lora16", "gpt2-psychia"],
    loss_type="trainloss",
    filename="graphs/lora_rank_finetuning_train.jpg",
    names=["Rank 8", "Rank 16", "Rank 32"],
    title="LoRA Rank Parameter Finetuning: Training Loss"
)

plot_curves(
    runs=["lora8", "lora16", "gpt2-psychia"],
    loss_type="evalloss",
    filename="graphs/lora_rank_finetuning_eval.jpg",
    names=["Rank 8", "Rank 16", "Rank 32"],
    title="LoRA Rank Parameter Finetuning: Validation Loss"
)
```
